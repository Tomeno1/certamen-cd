```python
#Importamos librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from apyori import apriori
```

Ejercicio 1:

```python
#Cargando los datos
df = pd.read_csv('breast-cancer.csv')
```

```python
#Mostrando Dataset antes
display(df.head(21))
```

```python
print("Valores ya reemplazados con NaN")

#Reemplazamos el caracter "?" por NaN
df.replace('?', pd.NA, inplace=True)

#Mostramos el Dataset
display(df.head(21))
```

```python
#Gracias al reemplazo del caracter "?" a NaN, se pudo determinar la cantidad de valores faltantes
df_faltantes = df.isnull().sum()
print("Los valores faltantes son los siguientes: ")
print(df_faltantes)


```

```python
#TENICA DE MINERIA DE DATOS, IMPUTACION POR LA MODA

# Variables categoricas
cols = ['node-caps', 'breast-quad']

# Guardando el df para visualizarlo después sin ningún cambio
df_original = df.copy()

# Imputación por la moda
imputer = SimpleImputer(strategy='most_frequent')

# Realizamos la imputación y reemplazamos las columnas originales
df[cols] = imputer.fit_transform(df[cols].astype('category'))

# Crea nuevas columnas con los valores imputados
imputed_cols = [f'{c}_imp' for c in cols]
df[imputed_cols] = df[cols]

#DataFrame original con valores nulos al lado del DataFrame con columnas imputadas
display(pd.concat([df_original[cols].head(21), df[imputed_cols].head(21)], axis=1, keys=['Original', 'Imputado']))


```

Ejercicio 2:


```python

# Carga el dataset de cáncer de mama
df = df_original

# Selecciona las columnas categóricas relevantes para el análisis
categorical_columns = ['age', 'menopause', 'tumor-size', 'inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad', 'irradiat', 'Class']

# Calcular la frecuencia de cada categoría en cada columna
frequencies = {}

for col in categorical_columns:
    frequencies[col] = df[col].value_counts()

# Identificar las categorías raras como posibles outliers
candidatos_outliers = {}

for col, freq in frequencies.items():
    candidatos_outliers[col] = freq[freq < 5]  # Se considera como outliers categorías con menos de 5 ocurrencias

# Mostrar los candidatos a outliers
for col, outliers in candidatos_outliers.items():
    print(f"Candidatos a outliers en la columna '{col}':")
    print(outliers)
    print()
```

Ejercicio 3:

```python
# Crear el df con los datos
data = df_original

df = pd.DataFrame(data)

#Convertir las variables categóricas a numéricas utilizando one-hot encoding
df_encoded = pd.get_dummies(df.drop('Class', axis=1))

#Crear el modelo con el metodo del arbol
model = RandomForestClassifier()

#Entrenar el modelo con tus datos
model.fit(df_encoded, df['Class'])

#Obtener la importancia de características
importancias = model.feature_importances_

#Crear un DataFrame con la importancia de características
feature_importance = pd.DataFrame({'Atributo': df_encoded.columns, 'Importancia': importancias})

#Ordenar los atributos por su importancia descendente
feature_importance = feature_importance.sort_values(by='Importancia', ascending=False)

#Imprimir los resultados
print(feature_importance)
```

```python
#importing the weather dataset 
url = 'breast-cancer.csv'
weather = pd.read_csv(url)
num_records = len(weather)
num_columns = len(weather.columns)

print("Cantidad de instancias: ", num_records)
print("Cantidad de columnas: ", num_columns)
print(weather)
```

```python
#modifying the weather dataset
records=[]
for i in range(0, num_records):
  records.append([weather.columns[j]+"="+str(weather.values[i,j]) for j in range(0, num_columns)])
print(records)
```

```python
my_association_rules = apriori(records,min_confidence=0.9)
association_results = list(my_association_rules) #best format to show rules
```

```python
rule_number = 0
for item in association_results:

    items_base  = item[2][0][0]
    antecedente = [x for x in items_base]

    items_add   = item[2][0][1]
    concecuente = [x for x in items_add]
 
    rule_number += 1
    print("Rule #",rule_number,":  ",antecedente," --> ",concecuente)
    print("Support: " + str(item[1]))
    print("Confidence: " + str(item[2][0][2]))
    print("=====================================")
```
