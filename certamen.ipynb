```python
#Importamos librerias necesarias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.impute import SimpleImputer

```

```python
#Cargando los datos
df = pd.read_csv('breast-cancer.csv')
```

```python
#Mostrando Dataset antes
display(df.head(21))
```

```python
print("Los valores faltantes")

#Reemplazamos el caracter "?" por NaN
df.replace('?', pd.NA, inplace=True)

#Mostramos el Dataset
display(df.head(21))
```

```python
#Gracias al reemplazo del caracter "?" a NaN, se pudo determinar la cantidad de valores faltantes
df_faltantes = df.isnull().sum()
print("Los valores faltantes son los siguientes: ")
print(df_faltantes)


```

```python
#TENICA DE MINERIA DE DATOS, IMPUTACION POR LA MODA

# Variables categoricas
cols = ['node-caps', 'breast-quad']

# Guardando el df para visualizarlo después sin ningún cambio
df_original = df.copy()

# Imputación por la moda
imputer = SimpleImputer(strategy='most_frequent')

# Realizamos la imputación y reemplazamos las columnas originales
df[cols] = imputer.fit_transform(df[cols].astype('category'))

# Crea nuevas columnas con los valores imputados
imputed_cols = [f'{c}_imp' for c in cols]
df[imputed_cols] = df[cols]

#DataFrame original con valores nulos al lado del DataFrame con columnas imputadas
display(pd.concat([df_original[cols].head(21), df[imputed_cols].head(21)], axis=1, keys=['Original', 'Imputado', 'PEDOs']))


```
